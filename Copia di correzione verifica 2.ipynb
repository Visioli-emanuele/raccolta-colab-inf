creare una funzione che permetta di incrementare la variabile x di 10 volte in una lista di numeri random 

```python
import random
lst = []
lstm = []
for i in range(20):
  lst.append (random.randint(0, 100))
for z in range (20):
  x= lst[z]
  y = lambda x : x + 10
  lstm.append(y)
print (lst)
print (lstm)
```
