```python
import pandas as pd
df= pd.read_excel("https://github.com/wtitze/3E/blob/main/Medals.xlsx?raw=true")
df
```

```python
print("eliminare le colonne:", " Rank","Total","Rank by Total")
df.drop(columns=["Rank","Total","Rank by Total"],inplace=True)
df
```

```python
print("rinominare la colonna: ","Team/NOC	senza NOC")
df=df.rename(columns={"Team/NOC":"Team"})
df
```

```python
print("inserire una nuova colonna con il totale delle medaglie per nazione")
df["Total"]=df["Gold"]+df["Silver"]+df["Bronze"]
df
```

```python
print("trovare le nazzioni che hanno più di 80 medaglie")
df[df["Total"]>80][["Team/NOC"]]
```

```python
print("numero medio di medaglie")
df["Total"].mean()
```

```python
df[df["Total"]>df["Total"].mean()][["Team/NOC"]]
```
