```python
df=pd.read_csv("https://raw.githubusercontent.com/italia/covid19-opendata-vaccini/master/dati/consegne-vaccini-latest.csv")
df
```

```python
data= df.groupby("nome_area").sum().sort_values("numero_dosi", ascending=False)
data
names= list(data.index) #creao una lista delle regioni prendendolo dall'indice che da quello normale (1,2,3,...) si è spostao alla colonna su cuoi abbiamo fatto il raggruppamento (in questo caso "nome_area") e li trasformo in lista perché per il grafico ci serve una lista
values= list(data["numero_dosi"]) #creo una lista con il numero delle dosi
#max_value= data.sort_values("numero_dosi", ascending=False)["numero_dosi"].head(1)[0]
```

```python
#metodo 1:
fig= plt.figure() #creo una tela bianca 
ax= plt.axes() #con axes() disegno gli assi
ax.bar(names,values) #stampo il grafico
#"le7" ci dice la notazione di scentifica che ci dice che ci sono sette "0" prima dei numeri
```

```python
#metodo 2: 
#il metodo subplots() senza parametri restituisce una figura e un singolo asse 
fig,ax=plt.subplot()
ax.bar(names,values)
```

```python
fig.set_figwidth(12)
fig.set_figheight(4)
fig.autofmt_xdate(rotation=90)
ax.set_ylim([0,20000000])
ax.set_yticklabels(np.arange(0,20000001, step=2000000), fontsize=20)
ax.set_yticks(np.arange(0,20000001,step=2000000))
ax.set_title("Distribuzione dosi per regione", fontsize=20)
ax.set_xlabel("Regioni", fontsize=20)
ax.set_ylabel("Dosi di vaccino", fontsize=20)
ax.set_xticklabels(names,fontsize=20)
```
