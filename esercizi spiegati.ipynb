3. scrivere una funzione chiamata costruisciLista che dati due numeri interi restituisca la lista formata da tutti i numeri interi compresi tra i due numeri dati. Utilizzare poi la funzione in un programma che acquisisca dall'utente due numeri e visualizzi la lista formata da tutti i numeri interi compresi tra i due numeri dati ma a partire dal più grande fino al più piccolo.

```python
#chiedo 2 numeri all'utente
n1=int(input("inserire il numero da qui iniziare: "))
n2=int(input("inserire il numero finale: "))
#inizializzo la funzione
def costrusciLista (n1,n2):
  #creo la lista per inserire tutti i numeri tra n1 e n2
  lst3=[]
  #uso un ciclo "for" e gli dico:
  #cicla da i (che per per le varaibili nel "rage" prende il vaore di n1) fino a n2 e gli diciamo di ciclare 1 numero alla volta (grazie al ",1")
  #in range(n1,n2,1)= significa che "i" prende il valore della prima variabile (n1) e cicla aumentadno la "i" fino a che non diventa uguale a "n2" 
  for i in range(n1,n2,1):
    #diciamo al programma di aggiungere a "last3" il valore della "i"
    lst3.append(i)
    #resituiamo "last3"
  return lst3  
bello= costrusciLista(n1,n2)
#inveritamo la lista (esemprio: lista1= 1, 2, 3, 4. dopo il "reverse()" lista1 diventa 4, 3, 2, 1. quindi il primo diventa l'ultimo il penultimo il secondo e così via)
bello.reverse()
print(bello)
```

4) scrivere una funzione chiamata valoreAssoluto che dato un numero restituisca il suo valore assoluto. Utilizzare poi questa funzione per determinare il valore assoluto degli elementi di una lista inserita dall'utente.


```python
#inizializiamo la funzione
def valoreAssoluto(n):
  #controllo che il numero non sia positivo (perché un numero assoluto è il corrispondete positivo di ogni numero negativo)
  if n < 0:
    #trasformo il numero negativo in positvo
    res = n*-1
    return res
  #se il numero è positivo allora restituisco il numero 
  else:
    return n
#crea una lista con "list()", con "map()" faccio sì che tutto quello dentro le parentesi sia rifatto in tutta la lista 
#"int, input()" fa sì che si inserisca un numero inserito dall'utente e "split()" divide tutto quello che inseriamo separando i valori con gli spazzi
L = list(map(int, input().split()))
#faccio si che la lista "c" si riempia tutta (con il map()) di numeri assoluti perché 
#facendo "map(valoreAssoluto,l)" fa sì che per tutti i valori della lista "l" si richiami la funzione "valoreAssoluto" e quindi che poi lo metta nella lista "c"
c = list(map(valoreAssoluto,L))
print(c)
```
