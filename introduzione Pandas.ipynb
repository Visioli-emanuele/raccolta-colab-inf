```python
df[df["Age"]>35] #controllo se il dato nella colonna (età) sia maggiore di un altro numero (35)
```

```python
df[(df["Age"] > 35) & (df.Age < 40)].sort_values(["Sex","Age"],ascending= True)[["Name","Sex","Age"]] #controllo se il dato nella colonna (età) sia maggiore di un altro numero (35) e minore di un'altro (40)
#sort_valuesmette in ordine secondo i paramentri nelle "()" ascending per dire se crescente (True) o decerescente (False) 
#per stampare tutto bisogna scrivere [[e le colonne]] (doppie [] perché l'esterna ci diche che vogliamo predndere delle colonne mentre la [] interna serve per creare la lista delle colonne da prendere)
```

```python
df[df.Sex == "male"]["Age"].mean()#prende 1 colonna e cerca il valore e poi nella 2° [] gli diciamo dove deve fare il cacolo e .mean() fa la media 
```

```python
df[df.Sex == "female"]["Age"].mean()
```

```python
df.groupby("Sex").mean()[["Age"]] #prendere la colonna tra () come riferimento e fa la media su tutte le altre ma stampo solo l'età grazie a [["Age"]]
```

```python
df.groupby("Age").count() #filtrimao per l'età
```
