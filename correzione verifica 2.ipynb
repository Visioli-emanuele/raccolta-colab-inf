caricare una lista di 10 città e ordinarle per lunghezza decrescente 

```python
citta = ["roma","como","bra", "bari", "milano", "pisa", "varese", "lecco"]
cittaCrescente= sorted(citta,key=len)
print (cittaCrescente)
cittaDecrescente= sorted(citta,key=len,reverse= True)
cittaDecrescente2= cittaCrescente[:]
cittaDecrescente2= cittaCrescente.reverse()
print (cittaDecrescente)
print (cittaDecrescente2)
print ( cittaCrescente)
```
