far inserire all'utente una lista e dire se ci sono doppioni

```python
my_list= [int(n) for n in input("inserisci i numeri maggiori di: ").split()]

seen = set()
duplicate_item = [x for x in my_list if x in seen or (seen.add(x) or False)]

if not duplicate_item:
  print("tutto a posto")
else:
  print("beccato!")
```
