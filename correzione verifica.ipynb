data una frase se è troppo corta o troppo lunga

````python
frase= "data una frase visualizzare le ultime 3 parole "
listaParole=frase.split()
if len(listaParole)>=10 :
  print('troppo lunga')
elif len(listaParole)<=5:
  print('troppo corta')
else:
  print ('ok')
````
